```python
import pandas as pd
import numpy as np
```

```python
situcion_empleo = pd.read_csv('./data/otros/sit_laboral_distrito.csv')
```

```python
situcion_empleo
```

```python
df1 = pd.get_dummies(situcion_empleo, columns=['Situación laboral'])
```

```python
df1.head()
```

```python
df2 =df1.groupby(['Codi_Districte']).agg(
    laboral_desempleado = ('Situación laboral_Desempleado', 'sum'),
    laboral_ocupado = ('Situación laboral_Ocupado', 'sum'),
    laboral_otras_situaciones = ('Situación laboral_Otras_situaciones','sum')
)
df2.reset_index(inplace=True)
```

```python
df2.to_json('situacion_empleo.json', orient='records', lines=True)
```
